```python
from bs4 import BeautifulSoup as bs
import requests
import random
import time

for i in range(1,4):
    url = 'https://www.goodreads.com/list/show/86.Best_Children_s_Books?page=' + str(i)
    response  = requests.get(url)
    status = response.status_code
    if status == 200:
        page = response.text
        soup = bs(page)
    else:
        print(f"Oops! Received status code {status} for {url}")

    table = soup.find('table', class_="tableList")
    rows = [row for row in table.find_all('tr')] 
    book_data = []
    for row in rows:
        time.sleep(.5+2*random.random())
        book_list = details(row)
        headers = ['book_name', 'author_name',
                'book_rating', 'no_rating', 'score','no_votes',"author_country","author_dob","author_average_rating",
                "authot_no_rating","author_no_review","author_distinct_work","author_followers"]

    
        book_dict = dict(zip(headers, book_list)) 
        book_data.append(book_dict)
        print(book_dict)


    
```

```python
import pandas as pd
book_info = pd.DataFrame(book_data)  #convert list of dict to df
book_info.set_index('book_name', inplace=True)

book_info
```

```python
##Saving into a csv file after parsing 3 full pages ie 300 records
#Merging all of them for analysis in next notebook for Model Development and EDA
#totally parsed 18*2 Pages parsing 1800Records

book_info.to_csv("Children_Page1_2_3.csv")
```

```python
def details(book):
    
    list1 = []
    book_name = book.find(class_ ="bookTitle").text.replace('\n','')
    
    author_name = book.find(class_="authorName").text
    author_link = book.find(class_ = "authorName")["href"]
    
    #Dependant Variable
    try:
        book_rating = float(book.find(class_ = "minirating").text.strip().split('‚Äî')[0].split(" ")[0])
    except:
        book_rating = 0
   
    try:
        no_rating = int(book.find(class_ = "minirating").text.strip().split('‚Äî')[1].strip().split(' ')[0].replace(',',''))
    except:
        no_rating = 0
    
    try:
        score = int(book.find("span", class_ = "smallText uitext").text.strip().split('\n')[0].split(':')[1].strip().replace(',',''))
    except:
        score = 0
    
    try:
        no_votes = int(soup.find("span", class_ = "smallText uitext").text.strip().split('\n')[2].strip().split(' ')[0].replace(',',''))
    except:
        no_votes = 0
    #Author Features
    
    author_details = find_author_details(author_link)
    if len(author_details) == 0:
        list1 = [book_name,author_name,book_rating, no_rating,score,no_votes,author_details[0],author_details[1],
             author_details[2],author_details[3],author_details[4],author_details[5],author_details[6]]
    else:
        list1 = [book_name,author_name,book_rating, no_rating,score,no_votes,0,0,
             0,0,0,0,0]
    return list1
```

```python
#Developed a function to parse 18 pages of Top Children Book's Authors collecting author average rating 
#and other features 
#Error Handling in case issue while capturing the variables
```

```python
def find_author_details(author_link):

     response  = requests.get(author_link)
     status = response.status_code
     #print(status)
     if status == 200:
        page = response.text
        soup2 = bs(page)
     else:
        print(f"Oops! Received status code {status}")
        return []
        
    #4th Feature  Country of Birth of Author
     try:
        if (soup2.find(class_ ="dataTitle").text == "Born"):
            country = soup2.find(class_ ="dataTitle").next.next.strip().split(',')[-1].strip()
        else:
            country = ' '
     except:
        country = ' '
    
     #Feature Birth Year of Author
     try: 
         year = birth_year(soup2)
     except:
         year = 0
        
     data = (soup2.find("div", class_ = "hreview-aggregate").text)
     list = data.strip().replace('\n','').replace('¬∑','').replace('‚Ä¢','').split()
      #Feature Average rating of Author
     try:
         average_rating =  float(list[-10].replace(',',''))
     except:
         average_rating =0
      # Feature No of Rating
     try:
         no_rating = int(list[-9].replace(',',''))
     except:
         no_rating = 0
       # Feature No of Review
     try:
         no_review = int(list[-7].replace(',',''))
     except:
         no_review = 0
    
       # Feature Distnict work
     try:
         distinct_work = int(list[-5].replace(',',''))
     except:
         distinct_work = 0
       # Feature No of Followers   
     try:
         followers = int(soup2.find(class_ ="h2Container gradientHeaderContainer").text.split('(')[1].replace(')','').replace(',',''))
     except:
         followers = 0
   
     
     author_details = [country,year,average_rating,no_rating,no_review,distinct_work,followers]
     return (author_details)
```
